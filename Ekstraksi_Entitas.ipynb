```python
import re
```

```python
# Fungsi untuk ekstraksi lokasi, entitas, dan tag dari file
def extract_entities_from_file(file_path):
    # Regex untuk menemukan entitas dengan tag "PN" beserta lokasi dan tag
    pn_regex = r"\((\d+:\d+:\d+:\d+)\)\s([^\s]+)\sPN.*"

    # Menyimpan hasil ekstraksi dalam bentuk list of tuples (lokasi, entitas, tag)
    extracted_data = []

    # Membuka file untuk dibaca
    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            # Mencocokkan pola dengan regex
            match = re.search(pn_regex, line)
            if match:
                # Menyimpan lokasi, entitas, dan tag
                location = match.group(1)
                entity = match.group(2)
                tag = "PN"
                extracted_data.append((location, entity, tag))

    return extracted_data

# Path file teks yang ingin diproses
file_path = '/content/quranic-corpus-morphology-0.4.txt'

# Ekstraksi entitas
data = extract_entities_from_file(file_path)
```

```python
# Menampilkan hasil ekstraksi
print("Lokasi, Entitas, dan Tag yang ditemukan:")
for location, entity, tag in data:
    print(f"Lokal: {location}, Entitas: {entity}, Tag: {tag}")
```

```python
# Menyimpan hasil ke dalam file TSV
def save_to_tsv(data, output_file):
    with open(output_file, mode='w', encoding='utf-8') as file:
        # Menulis header (opsional)
        file.write("Lokasi\tEntitas\tTag\n")
        # Menulis data dengan tab sebagai pemisah
        for location, entity, tag in data:
            file.write(f"{location}\t{entity}\t{tag}\n")
```

```python
# Menyimpan hasil ke file TSV
save_to_tsv(data, 'extracted_entities.tsv')

print("Data telah disimpan dalam file 'extracted_entities.tsv'.")
```
